## Create, Compile, and Train the Model

```python
num_columns = X_train_scaled.shape[1]
num_columns
```

```python
# Find the number of columns in the X training data
number_input_features = len(X_train.columns)
# Create the input layer
inputs = layers.Input(shape=(num_columns,))

# Create at least two shared layers
shared_layer1 = layers.Dense(64, activation='relu',name='shared1')(inputs)
shared_layer2 = layers.Dense(128, activation='relu',name='shared2')(shared_layer1)
```

```python
# Create a branch for Department
# with a hidden layer and an output layer
department_hidden = layers.Dense(32, activation='relu',name='department_hidden')(shared_layer2)
department_output = layers.Dense(y_train_encoded_department.shape[1], activation='softmax', name='department_output')(department_hidden)
# Create the hidden layer


# Create the output layer

```

```python
# Create a branch for Attrition
# with a hidden layer and an output layer

# Create the hidden layer


# Create the output layer
attrition_hidden = layers.Dense(32, activation='relu',name='attrition_hidden')(shared_layer2)
attrition_output = layers.Dense(2, activation='softmax', name='attrition_output')(attrition_hidden)

```

```python
# Create the model
model = Model(inputs=inputs, outputs=[department_output, attrition_output])

# Compile the model

model.compile(optimizer='adam',
              loss={'department_output': 'categorical_crossentropy', 'attrition_output': 'binary_crossentropy'},
              metrics={'department_output': 'accuracy', 'attrition_output': 'accuracy'})

# Summarize the model
model.summary()
```

```python
# Train the model
model.fit(
    X_train_scaled,
    {'department_output': y_train_encoded_department, 'attrition_output': y_train_encoded_att},
    epochs=100,
)
```

```python
# Evaluate the model with the testing data
test_results = model.evaluate(X_test_scaled,   {'department_output': y_test_encoded_department, 'attrition_output': y_test_encoded_att},)
test_results
```
